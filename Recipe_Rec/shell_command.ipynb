### 레시피에 카테고리 추가

```python
from main_page.models import Recipe, Tag, Category1, Category2, Category3, Category4
import pandas as pd 
from tqdm import tqdm
import os
path = '/Users/bagjunhyeong/Recipe/recipe_/recipe_Data'
recipe = pd.read_csv(os.path.join(path, 'recipe.csv'))
ingredeint = pd.read_csv(os.path.join(path, 'ingredient.csv'))
content = pd.read_csv(os.path.join(path, 'content.csv'))
import random
random.seed(42)
l = list(range(recipe.recipe_id.nunique()))
random.shuffle(l)
l = l[:10000]

recipe = recipe[recipe.index.isin(l)]

recipe2idx = {j:i for i,j in enumerate(recipe.recipe_id.unique())}
content = content[content.recipe_id.isin(recipe.recipe_id.unique())]

for r,i in tqdm(recipe2idx.items()):
    cat_1 = recipe[recipe.recipe_id == r].cat1.values[0]
    cat_2 = recipe[recipe.recipe_id == r].cat2.values[0]
    cat_3 = recipe[recipe.recipe_id == r].cat3.values[0]
    cat_4 = recipe[recipe.recipe_id == r].cat4.values[0]

    c1 = Category1.objects.get(name=cat_1)
    c2 = Category2.objects.get(name=cat_2)
    c3 = Category3.objects.get(name=cat_3)
    c4 = Category4.objects.get(name=cat_4)

    rcp = Recipe.objects.get(recipe_id=i)
    
    rcp.category1 = c1
    rcp.category2 = c2
    rcp.category3 = c3
    rcp.category4 = c4
    rcp.save()
```


### 콘텐트 추가

```python
from main_page.models import Recipe, Tag
import pandas as pd 
from tqdm import tqdm
import os
path = '/Users/bagjunhyeong/Recipe/recipe_/recipe_Data'
recipe = pd.read_csv(os.path.join(path, 'recipe.csv'))
ingredeint = pd.read_csv(os.path.join(path, 'ingredient.csv'))
content = pd.read_csv(os.path.join(path, 'content.csv'))
import random
random.seed(42)
l = list(range(recipe.recipe_id.nunique()))
random.shuffle(l)
l = l[:10000]

recipe = recipe[recipe.index.isin(l)]

recipe2idx = {j:i for i,j in enumerate(recipe.recipe_id.unique())}
content = content[content.recipe_id.isin(recipe.recipe_id.unique())]

for r,i in tqdm(recipe2idx.items()):
    rcp = Recipe.objects.get(recipe_id=i)
    
    rcp.content = ''
    rcp.content += '<div class="ing-container">\t'
    rcp.content += '<div style="font-weight:600;" class="mb-3"><i class="check-title fa-solid fa-check"></i> 사용 재료 </div>\t'                   
    
    rcp.content += '<ur>\t'                        
    for k,v in ingredeint[ingredeint.recipe_id==r][['ingredient_key', 'ingredient_value']].values:
        rcp.content+= f'<li class="ing-txt">{k} : {v}</li>\t'
    
    rcp.content += '</ul>\t'    
    rcp.content += '</div>\t'
    rcp.content += '<hr class="line">'
    
    for n, (txt, img_url) in enumerate(content[content.recipe_id==r][['content_value', 'content_image']].values):
        if type(txt)==str and type(img_url) == str:         
            rcp.content += '<div class="d-flex flex-column">\t'
            rcp.content += f'<div class = "d-flex flex-row">\t'
            rcp.content += f'<div><span class="badge-step badge badge-pill"><i class="check-step fa-solid fa-check"></i>Step.{n+1}</span></div>\t'
            rcp.content += f'<div><h5 class="content-text">{txt}</h5></div>\t'
            rcp.content += f'</div>\t'
            rcp.content += f'<img src="{img_url}" class="content-img">\t'            
            rcp.content += '</div>\t'
                            
    rcp.save()
```

### 레시피&재료 등록

```python
from main_page.models import Recipe, Tag
import pandas as pd
from tqdm import tqdm
import os

path = '/Users/bagjunhyeong/Recipe/recipe_/recipe_Data'
recipe = pd.read_csv(os.path.join(path, 'recipe.csv'))
ingredeint = pd.read_csv(os.path.join(path, 'ingredient.csv'))

import random
random.seed(42)
l = list(range(recipe.recipe_id.nunique()))
random.shuffle(l)
l = l[:10000]

recipe = recipe[recipe.index.isin(l)]

recipe2idx = {j:i for i,j in enumerate(recipe.recipe_id.unique())}

for i in ingredeint.ingredient_key.unique():
    Tag(name=i, slug=i).save()

for recipe_id, hit, img_url, title, hook, serving, time, grade, _, _, _, _ in tqdm(recipe.values):
    Recipe(recipe_id=  recipe2idx[recipe_id], 
            title=title, 
            hook = hook,             
            img_url= img_url,
            count_hit = hit,
            serving = serving,
            elapsed_time = time,
            grade = grade
            ).save()

for r_name, r_id in tqdm(recipe2idx.items()):
    r = Recipe.objects.get(recipe_id=r_id)
    ings = ingredeint[ingredeint.recipe_id==r_name].ingredient_key.values
    for i in ings:
        r.ingredient.add(Tag.objects.get(name=i))
    
```
